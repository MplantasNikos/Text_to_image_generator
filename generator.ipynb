```python
from diffusers import AutoPipelineForText2Image
import torch
```

```python
pipe = AutoPipelineForText2Image.from_pretrained("stabilityai/sdxl-turbo", torch_dtype=torch.float16, variant="fp16",force_download=True, resume_download=False)
pipe.to("cuda")
```

```python
prompt = "a green mercendes a45 amg."
```

```python
image = pipe(prompt=prompt, num_inference_steps=1, guidance_scale=0.0).images[0]
```

```python
from PIL import Image  
import PIL
im1 = image.save("generated/image4.jpg") 
```
